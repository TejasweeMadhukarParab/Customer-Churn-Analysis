# Replacing blanks with 0 as tenure is 0 and no total charges are recorded

```python
df['TotalCharges'] = df['TotalCharges'].replace(" ", "0")
df['TotalCharges'] = df['TotalCharges'].astype("float")
```

```python
def conv(value):
    if value == 1:
        return "Yes"
    else:
        return "No"

df['SeniorCitizen'] = df['SeniorCitizen'].apply(conv)
```

```python
df.head(30)
```

# Converted 0 and 1 values of SeniorCitizen to yes/No to make it easier to understand

```python
ax = sns.countplot(data = df, x = "Churn", palette = ["blue","red"], hue = "Churn")
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.title("Count of Customers by Churn")
plt.show()
```

```python
gb = df.groupby("Churn").agg({"Churn":"count"})
gb
plt.figure(figsize = (5,5))
plt.pie(gb['Churn'], labels = gb.index, autopct = '%1.2f%%', colors =['blue', 'red'])
plt.title("Percentage of Churned Customers", fontsize = 15)
plt.show()
```

#from the given piechart we can conclude that 26.54% of our customers have churned out!
#now explore the reason behind it

```python
plt.figure(figsize = (5,5))
ax = sns.countplot(data = df, x = "gender", palette = ['blue', 'red'], hue = "Churn")
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])

plt.title("Churn by Gender")
plt.show()
```

```python
total_counts = df['SeniorCitizen'].value_counts()

# Count of Churn and Non-Churn within each SeniorCitizen group
senior_churn_counts = df.groupby(['SeniorCitizen', 'Churn']).size().unstack().fillna(0)

# Calculate percentages for each SeniorCitizen group and Churn category
senior_churn_percentage = senior_churn_counts.div(total_counts, axis=0) * 100

# Plot stacked bar chart
plt.figure(figsize = (5,5))
ax = senior_churn_percentage.plot(kind='bar', stacked=True, color=['blue', 'red'], figsize=(6, 4))

# Add percentage labels on top of the bars
for p in ax.patches:
    height = p.get_height()
    if height > 0:
        ax.annotate(f'{height:.1f}%', 
                    (p.get_x() + p.get_width() / 2., p.get_y() + height / 2.),
                    ha='center', va='center', fontsize=10, color='white')

plt.title("Churn by SeniorCitizen (%)")
plt.ylabel("Percentage of Total")
plt.xlabel("SeniorCitizen")
plt.show()
```

```python


# List of columns you want to create countplots for
columns = ['PhoneService', 'MultipleLines', 'InternetService',
           'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
           'TechSupport', 'StreamingTV', 'StreamingMovies']




# Create subplots
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(15, 15))  # 3x3 grid for 9 plots
axes = axes.flatten()  # Flatten the 2D axes array to 1D for easier indexing

# Loop through each column and create a countplot in the corresponding subplot
for i, column in enumerate(columns):
    sns.countplot(data=df, x=column, ax=axes[i], hue = "Churn", palette =['red','blue'])
    axes[i].set_title(f"Countplot of {column}")
    axes[i].set_ylabel("Count")
    axes[i].set_xlabel(column)

# Adjust layout to make sure plots fit nicely
plt.tight_layout()
plt.show()
```

customers who did not churn (No) are significantly higher across most categories. However, higher churn rates are observed among customers with fiber optic internet and those lacking OnlineSecurity or TechSupport.
